# Regression takes correlation to the next level
Regression provides an equation that will explain how much weight changes for each additional unit of height and to make predictions for specific heights.

## Terms
**Dependent Variable**: the variable you want to explain or predict. Often denoted with _y_

**Independent Variables**: the variables used to explain or predict changes in the dependent variable (also known as predictors, inputs). Often denoted with _X_

**Univariate / Multivariate**: univariate regressions are also known as simple regressions and use a single independent variable. Multivariate regressions use more than one independent variable.

**Coefficients**: describe the relationships between the independent and dependent variables

**P-value**: indicates whether the variables are statistically significant. Low p-values (typically <0.05) indicate the independent variable is statistically significant.

## Controlling for effects
A nice feature about regressions is that including a variable isolates the effects from that exposure.

For example, if you are studying the effect of coffee drinking on mortality, smoking could be a contributing effect. By including both amount of smoking and amount of coffee drinking as independent variables you isolate the role of each variable. The results of this model will show that coffee lowers the risk of mortalility while smoking increases it.

Note the importance of including relevant variables. Excluding smoking would cause the relationship for coffee drinking alone to be misestimated.
